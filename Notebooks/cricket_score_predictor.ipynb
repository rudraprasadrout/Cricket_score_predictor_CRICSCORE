```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("1st Innings T20.csv")
```

```python
df = df.drop(columns = ["Unnamed: 0","city","venue","tossWinner"])
```

```python
df.head()
```

```python
df.info()
```

```python
df.sample(10)
```

```python
df["matchId"] = df["matchId"].str.replace('_1','').astype("int64")
```

```python
 df.sample(5)
```

```python
df['current_score'] = df.groupby('matchId')['runs'].cumsum()
df['overs'] = df['balls'].astype(int)
df['ball_number'] = round((df['balls'] % 1) * 10).astype(int)
df['total_balls_bowled'] = (df['overs'] * 6) + df['ball_number']
df['CRR'] = (df['current_score'] / df['total_balls_bowled']) * 6
```

```python
df.sample(5)
```

```python
final_df = df.groupby('matchId').sum()['runs'].reset_index().merge(df,on = 'matchId')
```

```python
final_df['current_wickets'] = final_df.groupby('matchId')['player_out'].cumsum()
```

```python
final_df.head(5)
```

```python
final = final_df[['battingTeam','bowlingTeam','current_score','balls','CRR','current_wickets','runs_x']]
```

```python
final['wickets'] = final['current_wickets']
final['Total_runs'] = final['runs_x']
final = final.drop(columns = ['runs_x','current_wickets'])
```

```python
final.head()
```

```python
final = final.sample(frac=1).reset_index(drop=True)
```

```python
final.head()
```

```python
final.to_csv("Cricket_score_predictor.csv",index = False)
```
